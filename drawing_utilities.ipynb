```python
# Import libraries
from matplotlib import pyplot as plt
import numpy as np
```

```python
print(plt.style.available)
plt.style.use('fivethirtyeight')
# plt.style.use('seaborn')
# plt.style.use('seaborn-paper')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14,6))

ax[0].set_title("Gender")
gender = np.array([77, 36])
labels = np.array(['male', 'female'])
ax[0].pie(gender, startangle=90, labels=labels)
ax[0].legend()

ax[1].set_title("Age")
age = np.array(['19', '20', '21', '22', '23', '24', '26', '30'])
freq = np.array([4, 23, 51, 24, 8, 1, 1, 1])
ax[1].bar(age, freq)
plt.xticks(age)
ax[1].grid(False)

plt.tight_layout()
plt.show()
```
